### 여러개의 그래프 그리기

```python
# 1. figure 함수로 개체를 만들고
# 2. add_subplot 으로 그리려는 그래프 개수만큼 subplot 을 만든다. 
```

```python
flg = plt.figure()
ax1 = flg.add_subplot(2,1,1)       #2행 1열의 첫번째 그래프 구역
ax2 = flg.add_subplot(2,1,2)       #2행 1열의 두번째 그래프 구역
plt.show()
```

```python
flg = plt.figure()
ax1 = flg.add_subplot(1,3,1)
ax2 = flg.add_subplot(1,3,2)
ax3 = flg.add_subplot(1,3,3)
plt.show()
```

```python
import matplotlib.pyplot as plt            
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

x = range(0,100)
y = [v**2 for v in x]

ax1.plot(x,y)
ax2.bar(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
x
```

```python
sin_y = np.sin(x)
cos_y = np.cos(x)

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x, sin_y, "b-")
ax2.plot(x, cos_y, "r--")
plt.show()
```

```python
## 2행 2열 그래프 / sin, cos, tan, arctan
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
arctan_y = np.arctan(x)

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(x, sin_y)
ax2.plot(x, cos_y, "rs")
ax3.plot(x, tan_y, "c.")
ax4.plot(x, arctan_y, "yo")
plt.show()
```

### 라벨 및 범주 표시하기

```python
sin_y = np.sin(x)
cos_y = np.cos(x)

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x, sin_y, "b-")
ax2.plot(x, cos_y, "r--")

ax1.set_xlabel('x축')                     #축 이름 보이기 설정
ax1.set_ylabel('sin(x)')

ax2.set_xlabel('x축')                    ##한글 깨지는거는 추후에....
ax2.set_ylabel('cos(x)')
plt.show()
```

```python
# 범례 표시하기

x = np.arange(10)
plt.plot(x, 1*x, label='y = %ix' % 1)          #label로 범주 정하고, 마무리로 plt.legend()
plt.legend()
plt.show()
```

```python
plt.plot(x, 1*x, label = 'y=%ix' % 1)
plt.plot(x, 2*x, label = 'y=%ix' % 2)
plt.legend()
plt.show()
```

```python
x = np.arange(10)
for i in range(5):
    plt.plot(x, i*x, label = 'y = %ix' % i)
plt.legend(loc=4)
plt.show()
```

```python
## 범레의 위치

# upper left    2                /     upper center        9              /     upper right  1



# center left  6                  /    center           10              /        right    5       center right        7



# lower left    3                /     lower center         8            /      lower right         4
```

### pie chart 

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc         #한글 설정때문에 불러와야 함
from matplotlib import style
```

```python
#우분투 폰트 디렉토리 경로 /usr/share/fonts
font_name = font_manager.FontProperties(fname="/usr/share/fonts/truetype/nanum/NanumGothic.ttf").get_name()          #font 불러오기
rc('font', family=font_name)            # font 자원 할당 
style.use('ggplot')         #matplotlib에서 ggplot style 사용?? 
```

```python
pielabels = ['삼성전자', 'SK하이닉스', 'LG전자', '네이버', '카카오']
pieratio = [50, 20, 10, 10, 10]
plt.pie(pieratio, labels=pielabels)
plt.show()
```

```python
plt.pie(pieratio, labels=pielabels, shadow = True)       # 그림자 효과 주기
plt.show()
```

```python
plt.pie(pieratio, labels=pielabels, startangle = 90)       #startangle 파이 시작 각도 지정 
plt.show()
```

```python
import numpy as np
x = np.arange(0.0, 2* np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.figure()
plt.subplot(2,1,1)
plt.plot(x, sin_y, 'bs', label="sin")
plt.legend()
plt.subplot(2,1,2)
plt.plot(x, cos_y, 'ro', label = "cos")
plt.legend(loc = "lower right")
plt.ylabel('y 축정보')
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
z = np.mean(y)
plt.plot(x,y, "b", label="b.b")
plt.title("matplotlib 이용 그래프 그리기")
plt.xlabel('x축')
plt.ylabel('y축')
plt.text(3.5, 9, "평균 : " + str(z))             #plt.text(x좌표, y좌표, text)
plt.grid()
plt.legend()
plt.show()
```
